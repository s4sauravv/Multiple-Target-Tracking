```python
#Lets import necessory libreries

import cv2
import numpy as np

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
#Read the image nad convert into a grayscale

image=cv2.imread(r"C:\Users\TAR-TT060-Saurav K\Desktop\Untitled.png")
image=cv2.resize(image,None,fx=0.9,fy=0.9)
```

```python

#Lets convert in gray

gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow("Gray Scaler",gray)

cv2.waitKey(12000)

cv2.destroyAllWindows()
```

```python
#Now convert the grayscale image to binary image

ret,binary=cv2.threshold(gray,0,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

In the above code we converted the gray scale into binary. using THRESH_OTSU

```python
#Now detect the contours 

contours , hierarchy = cv2.findContours(binary,mode=cv2.RETR_TREE,method=cv2.CHAIN_APPROX_NONE)
```

```python
print(hierarchy)
```

```python
#Visualize the data structure 
print("Length of Contours{}",format(len(contours)))
print(contours)
```

```python
#Draw contours on th object image

image_copy=image.copy()
image_copy=cv2.drawContours(image_copy,contours,-1,(0,255,0),thickness=2)
image_copy
```

```python
#Visualizing the results
cv2.imshow('Gray Scale Image',gray)
cv2.imshow('Drawn Contours',image_copy)
cv2.imshow('Binary Image',binary)

cv2.waitKey(20000)
cv2.destroyAllWindows()
```
